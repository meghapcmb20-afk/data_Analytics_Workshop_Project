```python
data.columns=data.columns.str.replace(" ","_")
data
```

```python
data["Age"]=data["Age"].astype(int)
```

```python
data.drop_duplicates()
```

```python
data["Total_Marks"]=data["Assignment_Score"]+data["Internal_Marks"]+data["Final_Exam_Marks"]
```

```python
data["Total_Percentage"]=(data["Total_Marks"]/100)*100
data
```

```python
def grade(Marks):
  if Marks>=90:
    return "A"
  elif Marks>=80:
    return "B"
  elif Marks>=70:
    return "C"
  elif Marks>=60:
    return "D"
  else:
    return "F"
clean_data["Grade"]=data["Total_Percentage"].apply(grade)
data
```
